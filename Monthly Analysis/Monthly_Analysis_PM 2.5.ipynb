```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import math
```

```python
data_b_NO2=pd.read_excel("/Users/roshanchandraayyadevara/Documents/Dataset_Dessertation/Month/Mon_PM_2_5.xlsx",sheet_name='Birmingham')
data_l_NO2=pd.read_excel("/Users/roshanchandraayyadevara/Documents/Dataset_Dessertation/Month/Mon_PM_2_5.xlsx",sheet_name='London')
data_lp_NO2=pd.read_excel("/Users/roshanchandraayyadevara/Documents/Dataset_Dessertation/Month/Mon_PM_2_5.xlsx",sheet_name='Liverpool')
```

```python
df_b_NO2 = data_b_NO2.groupby('Year').agg({
    'Jan': 'mean',
    'Feb': 'mean',
    'Mar': 'mean',
    'Apr': 'mean',
    'May': 'mean',
    'Jun': 'mean',
    'Jul': 'mean',
    'Aug': 'mean',
    'Sep': 'mean',
    'Oct': 'mean',
    'Nov': 'mean',
    'Dec': 'mean'
}).reset_index()
```

```python
df_l_NO2 = data_l_NO2.groupby('Year').agg({
    'Jan': 'mean',
    'Feb': 'mean',
    'Mar': 'mean',
    'Apr': 'mean',
    'May': 'mean',
    'Jun': 'mean',
    'Jul': 'mean',
    'Aug': 'mean',
    'Sep': 'mean',
    'Oct': 'mean',
    'Nov': 'mean',
    'Dec': 'mean'
}).reset_index()
```

```python
df_lp_NO2 = data_lp_NO2.groupby('Year').agg({
    'Jan': 'mean',
    'Feb': 'mean',
    'Mar': 'mean',
    'Apr': 'mean',
    'May': 'mean',
    'Jun': 'mean',
    'Jul': 'mean',
    'Aug': 'mean',
    'Sep': 'mean',
    'Oct': 'mean',
    'Nov': 'mean',
    'Dec': 'mean'
}).reset_index()
```

# Birmingham's NO2

```python
import matplotlib.pyplot as plt

# Assuming 'Year' column is numeric, if it's not, you can convert it to a string using: df_b_NO2['Year'] = df_b_NO2['Year'].astype(str)

# Extract the monthly mean values for each month
data = [
    df_b_NO2['Jan'],
    df_b_NO2['Feb'],
    df_b_NO2['Mar'],
    df_b_NO2['Apr'],
    df_b_NO2['May'],
    df_b_NO2['Jun'],
    df_b_NO2['Jul'],
    df_b_NO2['Aug'],
    df_b_NO2['Sep'],
    df_b_NO2['Oct'],
    df_b_NO2['Nov'],
    df_b_NO2['Dec']
]

# Set up the figure and axis
plt.figure(figsize=(10, 6))
ax = plt.gca()

# Create the box plot with colors
box_plot = ax.boxplot(data, labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], patch_artist=True)

# Customizing the colors of boxes and median lines
box_colors = ['lightblue', 'lightgreen', 'lightcoral', 'tan', 'pink', 'lightyellow', 'lightgrey', 'lavender', 'lightskyblue', 'lightgreen', 'lightcoral', 'tan']
median_color = 'blue'

for patch, color in zip(box_plot['boxes'], box_colors):
    patch.set_facecolor(color)
    
for median in box_plot['medians']:
    median.set(color=median_color, linewidth=2)

# Adding labels, title, and legend
ax.set_xlabel('Month')
ax.set_ylabel('Monthly Mean')
ax.set_title('Monthly Mean PM 2.5 in Birmingham')

# Show the plot
plt.grid()
plt.show()
```

```python
import pandas as pd

data_without_column = df_b_NO2.drop(columns=['Year'])

# Calculate the summary statistics for the remaining columns
summary_stats = data_without_column.describe()

# Accessing specific description parameters using indexing
mean_value = summary_stats.loc['mean']  # Mean value for each column
std_deviation = summary_stats.loc['std']  # Standard deviation for each column
min_value = summary_stats.loc['min']  # Minimum value for each column
max_value = summary_stats.loc['max']  # Maximum value for each column

# Accessing specific description parameters using attribute access
mean_value = summary_stats.mean()  # Mean value for each column
std_deviation = summary_stats.std()  # Standard deviation for each column
min_value = summary_stats.min()  # Minimum value for each column
max_value = summary_stats.max()  # Maximum value for each column
variance= summary_stats.var()
print("Mean:")
print(mean_value)

print("\nStandard Deviation:")
print(std_deviation)

print("\nMinimum Value:")
print(min_value)

print("\nMaximum Value:")
print(max_value)

print("\nVariance:")
print(variance)
```

```python
sorted_mean = mean_value.sort_values(ascending=False)

# Print the sorted mean values
print("Sorted Mean Values (Highest to Lowest):")
print(sorted_mean)
```
